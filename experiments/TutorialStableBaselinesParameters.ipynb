#### Evaluation

- `mean_ep_length`: Mean Episode Length (seconds)
- `mean_reward`: Mean Episodic Reward


#### Rollout

The term "rollout" is in regard to an execution of a policy from the current state when there is some uncertainty about the next state or outcome, it is one simulation from the current state. It can also be defined as a simulated trajectory.

- `ep_len_mean`: Mean Episode Length (seconds)
- `ep_rew_mean`: Mean Episodic Reward


#### Time

- `episodes`: Total number of episodes
- `fps`: Number of frames per seconds or steps per second according to [this](https://github.com/hill-a/stable-baselines/issues/121#issuecomment-445469720)
- `iterations`: Number of iterations (data collection + policy update for A2C/PPO)
- `time_elapsed`: Time is seconds since the beginning of training
- `total_timesteps`: Total number of timesteps


Total timesteps is the number of steps in total the agent will do for any environment. The total timesteps can be across several episodes, meaning that this value is not bound to maximum.

According to [this](https://www.cfd-online.com/Forums/fluent/96834-time-step-iteration.html), timesteps are related to the real time of a simulation. Iteration is a cycle to reach a solution; in fact a timestep can contain multiple iterations.

$$ \text{number of updates} = \frac{\text{total timesteps}}{\text{batch size}} $$


##### What is the difference between epoch and episode?

- **Episode**: 
    - An episode is a sequence of states, actions, and rewards, which ends with a terminal state.
    
- **Epoch**: 
    - An epoch is one forward pass and one backward pass of all the training examples in neural network terminology. 
    - In other words, when all the data samples have been exposed to the neural network for learning patterns, one epoch is said to be completed.


#### Train

- `actor_loss`: Current value for the actor loss for off-policy algorithms

- `approx_kl`: approximate mean KL divergence between old and new policy (for PPO), it is an estimation of how much changes happened in the update

$$ \text{maximize}_{\theta} \ E_t \left[ \frac{\pi_{\theta}(a_t | s_t)}{\pi_{\text{θold}}(a_t | s_t)} \cdot A_t \right] $$
$$ \text{subject to} \ E_t \left[ \text{KL}[\pi_{\text{θold}}(\cdot | s_t), \pi_{\theta}(\cdot | s_t)] \right] \leq \delta $$


> **Description:**
In Trust Region Policy Optimization (TRPO), the algorithm maximizes an objective function known as the "surrogate" objective. This maximization is performed subject to a constraint on the size of the policy update.





- `clip_fraction`: mean fraction of surrogate loss that was clipped (above clip_range threshold) for PPO.

$$
L_{\text{CLIP}}(\theta) = E_t[\min(r_t(\theta)A_t, \text{clip}(r_t(\theta), 1 - \varepsilon, 1 + \varepsilon)A_t)]
$$

where

$$
r_t(\theta) = \frac{\pi_{\theta}(a_t | s_t)}{\pi_{\text{old}}(a_t | s_t)}
$$


- `clip_range`: Current value of the clipping factor for the surrogate loss of PPO. Roughly: how far can the new policy go from the old policy while still profiting. Typically denoted by ε.

- `critic_loss`: Current value for the critic function loss for off-policy algorithms, usually error between value function output and TD(0), temporal difference estimate

- `ent_coef`: Current value of the entropy coefficient (when using SAC)

- `ent_coef_loss`: Current value of the entropy coefficient loss (when using SAC)

> **Description:**
A central feature of SAC is entropy regularization. The policy is trained to maximize a trade-off between expected return and entropy, a measure of randomness in the policy. This has a close connection to the exploration-exploitation trade-off: increasing entropy results in more exploration, which can accelerate learning later on.

> In entropy-regularized reinforcement learning, the agent gets a bonus reward at each time step proportional to the entropy of the policy at that timestep. This changes the RL problem to:

$$
\pi^* = \arg \max_{\pi} \mathbb{E}_{\tau \sim \pi} \left[ \sum_{t=0}^{\infty} \gamma^t \left( R(s_t, a_t, s_{t+1}) + \alpha H\left(\pi(\cdot|s_t)\right) \right) \right]
$$

where $H$ is the entropy of $x$ computed from its distribution $P$ according to

$H(P) = -\sum_{x \sim P} \log P(x)$



- `entropy_loss`: Mean value of the entropy loss (negative of the average policy entropy)

> According to [this](https://fosterelli.co/entropy-loss-for-reinforcement-learning), entropy loss is an additional loss parameter that can be added to a model to help with local minima in the same way that momentum might and to provide encouragement for the agent to take a variety of moves and explore the set of strategies more.

> When we calculate the training loss at the end of the epoch, we’ll subtract the entropy loss over all of the timesteps from the reward-based policy and value losses.

> In the PPO algorithm, the loss function mentioned above is implemented in the code of the stable-baselines3 library. This can be seen [here](https://stable-baselines3.readthedocs.io/en/v1.0/_modules/stable_baselines3/ppo/ppo.html).

- `explained_variance`: Fraction of the return variance explained by the value function, see https://scikit-learn.org/stable/modules/model_evaluation.html#explained-variance-score (ev=0 => might as well have predicted zero, ev=1 => perfect prediction, ev<0 => worse than just predicting zero)

$$\text{explained variance} = 1 - \frac{\text{Var}[\text{empirical return} - \text{predicted value}]}{\text{Var}[\text{empirical return}]}$$

> Explained variance refers to the variance in the response variable in a model that can be explained by the predictor variable(s) in the model.

> The higher the explained variance of a model, the more the model is able to explain the variation in the data.

- `learning_rate`: Current learning rate value

- `loss`: Current total loss value

- `n_updates`: Number of gradient updates applied so far

- `policy_gradient_loss`: Current value of the policy gradient loss (its value does not have much meaning)

- `value_loss`: Current value for the value function loss for on-policy algorithms, usually error between value function output and Monte-Carlo estimate (or TD(lambda) estimate)

- `std`: Current standard deviation of the noise when using generalized State-Dependent Exploration (gSDE)
