## Find the maximum average reward for the AV agents

```python
import os
import pandas as pd

def process_av_rewards(directory_path, start_episode=201):
    """
    Process CSV files in the specified directory to find the maximum average reward
    of AV agents and their corresponding actions and IDs.

    Args:
    - directory_path (str): The path to the directory containing CSV files.
    - start_episode (int): The episode number to start processing from (default is 201).

    Returns:
    - max_average_reward (float): The maximum average reward found.
    - max_filename (str): The filename corresponding to the maximum average reward.
    - max_av_actions (list): The actions of AV agents corresponding to the maximum average reward.
    - max_av_ids (list): The IDs of AV agents corresponding to the maximum average reward.
    """
    
    # List to hold the CSV filenames
    csv_filenames = []

    # Variable to track maximum average reward and its corresponding filename and actions
    max_average_reward = float('-inf')  # Start with the lowest possible value
    max_filename = None
    max_av_actions = []
    max_av_ids = []  # List to store the IDs of AV agents

    # Iterate over all files in the specified directory
    for filename in os.listdir(directory_path):
        if filename.startswith("ep") and filename.endswith(".csv"):
            # Extract the episode number from the filename
            episode_number = int(filename[2:-4])  # Assuming the format is "ep<number>.csv"
            csv_filenames.append((episode_number, filename))

    # Sort the filenames based on the episode number
    csv_filenames.sort(key=lambda x: x[0])

    # Loop over the sorted filenames starting from the specified episode
    for episode_number, filename in csv_filenames:
        if episode_number >= start_episode:  # Start processing from ep201.csv
            file_path = os.path.join(directory_path, filename)
            
            # Read the CSV file into a DataFrame
            df = pd.read_csv(file_path)
            
            # Filter for AV agents and calculate average reward
            av_rewards = df[df['kind'] == 'AV']['reward']
            av_actions = df[df['kind'] == 'AV']['action']
            av_ids = df[df['kind'] == 'AV']['id']  # Get the IDs of AV agents
            
            if not av_rewards.empty:  # Check if there are any AV entries
                average_reward = av_rewards.mean()
                print(f"Processing {filename}: Average Reward of AV agents = {average_reward:.2f}")

                # Check if this average reward is the maximum found so far
                if average_reward > max_average_reward:
                    max_average_reward = average_reward
                    max_filename = filename
                    max_av_actions = av_actions.tolist()  # Store the actions as a list
                    max_av_ids = av_ids.tolist()  # Store the IDs as a list
            else:
                print(f"Processing {filename}: No AV agents found.")

    # After processing all files, print the maximum average reward and its corresponding filename and actions
    if max_filename is not None:
        print(f"\nMaximum Average Reward of AV agents: {max_average_reward:.2f} from file {max_filename}")
        print(f"Actions of AV agents in {max_filename}: {max_av_actions}")
        print(f"IDs of AV agents in {max_filename}: {max_av_ids}")
    else:
        print("\nNo AV agents found in any files.")
    
    return max_average_reward, max_filename, max_av_actions, max_av_ids  # Return the IDs as well
```
