```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read from csv file

df1 = pd.read_csv('ep1.csv')
df1.head()

# read from xml file

df2 = pd.read_xml('ep1.xml')
df2.head()
```

```python
#join the two dataframes on the column 'id'

df = pd.merge(df1, df2, on='id')
df.head()
```

```python
df["dep-delay"] = df["depart"] - df["departDelay"]

df["travelTime"] = (df["duration"] + df["departDelay"])/60

df["diff"] = df["travelTime"] - df["travel_time"]

print(df["diff"].describe())
```
